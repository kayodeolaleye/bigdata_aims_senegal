```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## a. This line reads in the data (UN.csv)

```python
df = pd.read_csv('data/UN.csv')
df
```

## b. Count the number of rows and columns in the data

```python
df.info()
```

## c. Plot and analysis of GDP vs Infant Mortality

```python
#Create a data containing GDP values and Infant Mortality values
GDPInfantMort = pd.DataFrame({'GDP': df.GDPperCapita, 'Infant Mortality':df.infantMortality}).dropna()
```

```python

#Before creating any figure we define what X and Y are
X = GDPInfantMort['Infant Mortality'].values
Y = GDPInfantMort['GDP'].values

#plot the figure using the "scatter" function
#plt.scatter(X,Y)
#Note that you can change the marker type via the "marker" keyword. For instance try :
plt.scatter(X,Y,marker='^')

# #We now give the same scale and limits to the x-axis and the y-axis
# plt.axis('scaled')
# plt.xlim(0.0,50000.0)
# plt.ylim(0.0,50000.0)

#Then let's create the y=x line
plt.plot([0.0,200.0],[0.0,50000.0],color='k')

#we can also add a grid
plt.grid()

#but more important we need to label our axes ...
plt.xlabel('Infant Mortality')
plt.ylabel('GDP')
#... and give a title (not necessary if you use a caption)
plt.title('GDP versus Infant Mortality')

```

### The above figure means that high GDP results in low infant mortality and vice versa

## d. Plot and analysis of Male Life Expectancy vs Female Life Expectancy

```python
#Create a data containing GDP values and Infant Mortality values
GDPInfantMort = pd.DataFrame({'Male Life Exp': df.lifeMale, 'Female Life Exp':df.lifeFemale}).dropna()
```

```python
#Before creating any figure we define what X and Y are
X = GDPInfantMort['Female Life Exp'].values
Y = GDPInfantMort['Male Life Exp'].values

#plot the figure using the "scatter" function
#plt.scatter(X,Y)
#Note that you can change the marker type via the "marker" keyword. For instance try :
plt.scatter(X,Y,marker='^')

#We now give the same scale and limits to the x-axis and the y-axis
plt.axis('scaled')
plt.xlim(0.0,90.0)
plt.ylim(0.0,90.0)

# Then let's create the y=x line
plt.plot([0.0,90.0],[0.0,90.0],color='k')

#we can also add a grid
plt.grid()

#but more important we need to label our axes ...
plt.xlabel('Female Life Exp')
plt.ylabel('Male Life Exp')
#... and give a title (not necessary if you use a caption)
plt.title('Male Life Expectancy vs Female Life Expectancy')
```

## e. Additional Plots

```python
Additionalplots_data = pd.DataFrame({'Economic Activity Male': df.economicActivityMale, 'Economic Activity Female':df.economicActivityFemale}).dropna()
plt.subplot(211)
plt.bar(Additionalplots_data['Economic Activity Male'].index, Additionalplots_data['Economic Activity Male'])
plt.grid()
plt.title('Economic Activity Male')
plt.ylabel('Economic Activity Male')

plt.subplot(212)
plt.bar(Additionalplots_data['Economic Activity Female'].index, Additionalplots_data['Economic Activity Female'])
plt.grid()
plt.title('Economic Activity Female')
plt.ylabel('Economic Activity Female')

plt.subplots_adjust(hspace = 0.5)
```
