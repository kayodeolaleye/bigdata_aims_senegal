```python
import numpy as np
all_guesses = []
participant_total = 50
randUpper = 100
#This function stores 50 random guesses in the list "all_guesses" and returns the list
def guessList():
    i = 0
    for i in range(participant_total):
        guesses = np.random.randint(1, randUpper)
        all_guesses.append(guesses)
        i+=1
    return "A list of all guesses excluding my guess: ", all_guesses

guessList()
```

```python
#This function stores my guess in a variable name "myguess" and append it to the list of all guesses
def myGuess(History=None):
    myguess = 50
    guessList()
    all_guesses.append(myguess)
    return "A list of all guesses including my guess", all_guesses
myGuess() 
```

```python
#This function computes the mean of all guesses
def meanGuess():
    myGuess()
    meanguess = np.mean(all_guesses)
    return "The mean of the guesses = ", meanguess
meanGuess()
```
